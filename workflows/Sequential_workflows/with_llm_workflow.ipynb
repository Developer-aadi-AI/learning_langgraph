```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_ollama import ChatOllama
from IPython.display import Image
```

```python
base_url = "http://localhost:11434"
model = "llama3.2"
llm = ChatOllama(
    base_url=base_url,
    model=model,
    temperature=0.5
)
```

```python
class QA_State(TypedDict):
    question: str
    answer: str
```

```python
graph = StateGraph(QA_State)
```

```python
def QA_System(state: QA_State)-> QA_State:
    question = state['question']
    prompt = f"Answer the following Question {question}"
    answer = llm.invoke(prompt).content
    state['answer'] = answer
    return state
```

```python
graph.add_node('QA_System', QA_System)

graph.add_edge(START, 'QA_System')
graph.add_edge('QA_System', END)

```

```python
workflow = graph.compile()
```

```python
initial_state = {'question' : "What is LangGraph?"}
final_state = workflow.invoke(initial_state)

print(final_state['answer'])
```
